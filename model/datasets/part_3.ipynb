```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
df1 = pd.read_csv("../datasets/P3_Data.csv")
df2 = df1.copy()
df2.head()
```

```python
import pandas as pd

# Categories for one-hot encoding
categories = [
    'Agricultural',
    'Commercial',
    'Residential',
    'Other'
]

# Define a function to apply one-hot encoding
def apply_one_hot_encoding(row):
    one_hot_encoding = [0] * len(categories)
    
    # Split the row string into categories
    row_categories = row.split(',')
    
    # Update one-hot encoding based on presence of categories
    for category in row_categories:
        if category in categories:
            index = categories.index(category)
            one_hot_encoding[index] = 1
    
    return one_hot_encoding

# Apply one-hot encoding to the 'Land_type' column
df_encoded = pd.DataFrame(df2['Land_type'].apply(apply_one_hot_encoding).tolist(), columns=categories)

# Concatenate the original DataFrame with the encoded one-hot columns
df3 = pd.concat([df2, df_encoded], axis=1)

# Drop the original 'Land_type' column
df3 = df3.drop('Land_type', axis=1)

df3.head()
```
